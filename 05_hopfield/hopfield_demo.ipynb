# Demo Rete di Hopfield - Pattern Recall

L'algoritmo dimostra la recall di un pattern corrotto.
L'utente deve scegliere un pattern di input e il livello di rumore,
e un intervallo in cui mostrare i progressi dell'algoritmo.

```octave
load hop_data
figure(1)
hop_plotpats(P)
```

Calcolo della matrice dei pesi:

```octave
W = hop_stor(P);
```

Scegli un pattern di partenza e il livello di rumore da usare per la corruzione:

```octave
% Pattern (numero da 1 a 8 corrispondente all'immagine)
pat = 1;
% Probabilità di flip (da 0 a 1)
pr = 0.3;
% Ogni quante iterazioni mostrare lo stato della rete
u = 5;



test_pat = hop_flip(P(:,pat),pr);
[s,c,M]=hop_test_pure(W,test_pat);

figure(1)
set(gcf,'Position',[11 653 645 244])
clf
subplot(1,3,1)
hop_plotdig(P(:,pat),12,10,'Input Pattern')


subplot(1,3,2)
hop_plotdig(test_pat,12,10,'Corrupted Pattern')


subplot(1,3,3)
hop_plotdig(s,12,10,'Final State');

figure(2)
set(gcf,'Position',[11 653 645 244])
fig_count = idivide(c, u) + 1 + (u != c);
columns = 5;
rows = idivide(fig_count, columns, 'ceil');

% subplot(rows, columns, 1)
% hop_plotdig(M(:, 1),12,10, 'Initial')

subplot_i = 0;
for i = [1, u:u:c]
    subplot_i += 1;
    subplot(rows, columns, subplot_i)
    hop_plotdig(M(:, i),12,10, sprintf('Iteration %d', i))
end

if i != c
    subplot(rows, columns, subplot_i + 1)
    hop_plotdig(M(:, c),12,10, sprintf('Iteration %d', c))
end

% if (i * u) != c
%     subplot(rows, columns, i + 1)
%     hop_plotdig(M(:, c),12,10, sprintf('Iteration %d', c))
% end
```
